```python
import pandas as pd 
import matplotlib.pyplot as plt
df=pd.read_csv('data.csv',index_col='Year',parse_dates=['Year'])
df.head()
df.Monaco=pd.to_numeric(df.Monaco,errors='coerce')
```

```python
df_totals=df.sum(axis=0).sort_values()
```

```python
df_totals
```

## Plotting Bar Graph

```python
ax=df_totals.plot(kind='bar',figsize=(10,5),title='CO$_2$ Emissions')
ax.set_ylabel('CO Emission')
ax.get_yaxis().get_major_formatter().set_scientific(False)
plt.show()
```

## Slicing

```python
df_totals=df.sum(axis=0).sort_values()[-5:]
ax=df_totals.plot(kind='bar',figsize=(10,5),title='CO$_2$ Emissions')
ax.set_ylabel('CO Emission')
ax.get_yaxis().get_major_formatter().set_scientific(False)
plt.show()
```
