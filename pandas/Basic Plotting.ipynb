```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,4*np.pi,1000)
volts=np.sin(x)+0.25*np.random.rand(len(x))
current=np.cos(x)+0.25*np.random.rand(len(x))

plt.plot(x,volts,label="volts")
plt.plot(x,current,label="current")
plt.legend()
plt.show()
```

# Using Dataframes

```python
signal_dict={
    'volts':volts,
    'current':current
}
signal_df=pd.DataFrame(signal_dict)
signal_df.head()
```

```python
signal_df.plot()
plt.show()
```

```python
signal_df['volts'].plot()
plt.show()
```

```python
signal_df[['volts','current']].plot()
plt.show()
```
