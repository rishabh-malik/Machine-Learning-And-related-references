# Rolling Filter

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib nbagg
```

```python
df=pd.read_csv('all_stocks_1yr.csv',index_col='Date',parse_dates=['Date'])
```

```python
df.head()
```

```python
df_aapl=df[df.Name=='AAPL']
```

```python
df_aapl.head()
```

```python
win_length=10
win_type=[None,'boxcar','triang','blackman','hamming','nuttall']

ax=df_aapl['Open'].plot(title='Apple Open price')

plt.show()
```

```python
df_aapl_roll=df_aapl.rolling(win_length).mean()
df_aapl_roll.plot()
plt.show()
```
